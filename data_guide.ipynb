```python
import numpy as np
import pandas as pd
```

```python
csv_file = 'dft-road-casualty-statistics-accident-2020.csv'
dataset_type = 'Accident' # can be one of ['Accident', 'Vehicle', 'e_scooter', 'Casualty']
df = pd.read_csv(csv_file, low_memory=False)
# df = pd.read_csv(csv_file, nrows=10000) # use this line if you want to only experiment with a small sample of the dataset
df
```

```python
df_metadata = pd.read_excel('dataguide.xlsx')
df_metadata = df_metadata.loc[df_metadata['table'] == dataset_type]
df_metadata = df_metadata[['field name', 'code/format', 'label']]
df_metadata
```

```python
# Create a table that contains the dataset columns as index and ALL the possible codes from ALL the columns of the dataset(thats why there are NaN values)
table = pd.pivot_table(df_metadata, values='label', index='field name', columns='code/format', aggfunc=max)
table
```

```python
# For each column of the dataset if column belongs to the columns that have been encoded according to dataguide.xlsx then replace the values/codes of the column with the labels
for col in df.columns:
    if col in table.index:
        dic = dict()
        for value in df[col].unique():
            if value in table.columns and not pd.isna(table.loc[col][value]):
                dic[value] = table.loc[col][value]
        df[col] = df[col].map(dic)

df
```
